```python
        #data visualizations

import matplotlib.pyplot as plt

# Scatter plot: Order Count vs. Purchase Amount
plt.scatter(customer_data["order_count"], customer_data["purchase_amount"])
plt.xlabel("Order Count")
plt.ylabel("Purchase Amount")
plt.title("Order Count vs. Purchase Amount")
plt.show()
```

```python
        #data visualizations

import matplotlib.pyplot as plt

# Calculate mean purchase amount for each order count
mean_purchase_by_order_count = customer_data.groupby("order_count")["purchase_amount"].mean()

# Bar plot: Mean Purchase Amount for Each Order Count
plt.bar(mean_purchase_by_order_count.index, mean_purchase_by_order_count.values)
plt.xlabel("Order Count")
plt.ylabel("Mean Purchase Amount")
plt.title("Mean Purchase Amount for Each Order Count")
plt.show()

```
