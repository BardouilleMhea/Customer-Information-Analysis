```python
        #data preparation

# import package to read csv file (pandas)
import pandas as pd

# read data from csv file
customer_data = pd.read_csv("customers.csv")

```

```python
        #data exploration

gender_counts = customer_data["gender"].value_counts()
print(gender_counts)
```

```python
        #data exploration -  grouping and aggregation
    
gender_grouped = customer_data.groupby("gender").agg( 
    purchase_amount_mean=("purchase_amount", "mean"),
    order_count_mean=("order_count", "mean"),
    average_order_value_mean=("average_order_value", "mean"))

gender_grouped = gender_grouped.round(2)

print(gender_grouped)

```

```python
        #data visualization 
    
import pandas as pd
import matplotlib.pyplot as plt

# Data visualization
# Creates a new figure with a specific size of 10 inches in width and 5 inches in height
plt.figure(figsize=(10, 5))

# Bar plot for Mean Purchase Amount
plt.subplot(1, 3, 1)
plt.bar(gender_grouped.index, gender_grouped["purchase_amount_mean"])
plt.xlabel("Gender")
plt.ylabel("Mean Purchase Amount")
plt.title("Mean Purchase Amount by Gender")

# Bar plot for Mean Order Count
plt.subplot(1, 3, 2)
plt.bar(gender_grouped.index, gender_grouped["order_count_mean"])
plt.xlabel("Gender")
plt.ylabel("Mean Order Count")
plt.title("Mean Order Count by Gender")

# Bar plot for Mean Average Order Value
plt.subplot(1, 3, 3)
plt.bar(gender_grouped.index, gender_grouped["average_order_value_mean"])
plt.xlabel("Gender")
plt.ylabel("Mean Average Order Value")
plt.title("Mean Average Order Value by Gender")

plt.tight_layout()
plt.show()
```

```python
# pie chart of gender distribution 

gender_distribution = customer_data['gender'].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(gender_distribution, labels=gender_distribution.index, autopct='%1.1f%%')
plt.title('Gender Distribution')
plt.show()
```
