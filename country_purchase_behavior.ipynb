```python
        #data preparation

# import package to read csv file (pandas)
import pandas as pd

# read data from csv file
customer_data = pd.read_csv("customers.csv")

```

```python
        #data exploration

country_counts = customer_data['country'].unique()
print(country_counts)
```

```python
        #data exploration -  grouping and aggregation
    
country_grouped = customer_data.groupby("country").agg( 
    age_mean = ("age", "mean"),
    purchase_amount_mean=("purchase_amount", "mean"),
    order_count_mean=("order_count", "mean"),
    average_order_value_mean=("average_order_value", "mean"))

country_grouped = country_grouped.round(2)

print(country_grouped)

```

```python
        #data visualization 
    
import pandas as pd
import matplotlib.pyplot as plt

# Group data by country and calculate mean purchase_amount
country_grouped = customer_data.groupby("country")["purchase_amount"].mean().reset_index()

# Sort the data by mean purchase_amount
country_grouped = country_grouped.sort_values(by="purchase_amount", ascending=False)

# Create a bar plot
plt.bar(country_grouped["country"], country_grouped["purchase_amount"])
plt.xlabel("Country")
plt.ylabel("Mean Purchase Amount")
plt.title("Mean Purchase Amount by Country")
plt.xticks(rotation=45)
plt.show()

```

```python
        #data visualization 
    
import pandas as pd
import matplotlib.pyplot as plt

# Group data by country and calculate mean order_count
country_grouped = customer_data.groupby("country")["order_count"].mean().reset_index()

# Sort the data by mean order_count
country_grouped = country_grouped.sort_values(by="order_count", ascending=False)

# Create a bar plot
plt.bar(country_grouped["country"], country_grouped["order_count"])
plt.xlabel("Country")
plt.ylabel("Mean Order Count")
plt.title("Mean Order Count by Country")
plt.xticks(rotation=45)
plt.show()
```

```python
        #data visualization 
    
import pandas as pd
import matplotlib.pyplot as plt

# Create a box plot
plt.boxplot(
    [customer_data.loc[customer_data["country"] == country, "purchase_amount"] for country in country_grouped["country"]],
    labels=country_grouped["country"]
)
plt.xlabel("Country")
plt.ylabel("Purchase Amount")
plt.title("Purchase Amount Distribution by Country")
plt.xticks(rotation=45)
plt.show()

```
