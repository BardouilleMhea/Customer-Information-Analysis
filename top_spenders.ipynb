```python
        #data preparation

# import package to read csv file (pandas)
import pandas as pd

# read data from csv file
customer_data = pd.read_csv("customers.csv")

```

```python
        #data visualization 
    
import pandas as pd
import matplotlib.pyplot as plt

# Bar Plot - Top Spenders by Purchase Amount
top_spenders = customer_data.sort_values(by='purchase_amount', ascending=False).head(10)
plt.bar(top_spenders['first_name'] + ' ' + top_spenders['last_name'], top_spenders['purchase_amount'])
plt.xlabel('Customer')
plt.ylabel('Purchase Amount')
plt.title('Top 10 Spenders by Purchase Amount')
plt.xticks(rotation= 50)
plt.subplots_adjust(bottom= 0.25)
plt.show()
```

```python
        #data visualization 
    
import pandas as pd
import matplotlib.pyplot as plt

# Scatter Plot - Purchase Amount vs. Order Count
plt.scatter(customer_data['purchase_amount'], customer_data['order_count'])
plt.xlabel('Purchase Amount')
plt.ylabel('Order Count')
plt.title('Purchase Amount vs. Order Count')
plt.show()
```

```python
        #data visualization 
    
import pandas as pd
import matplotlib.pyplot as plt

# Visualization 3: Box Plot - Purchase Amount by Country
plt.boxplot([customer_data[customer_data['country'] == country]['purchase_amount'] for country in customer_data['country'].unique()])
plt.xlabel('Country')
plt.ylabel('Purchase Amount')
plt.title('Purchase Amount by Country')
plt.xticks(range(1, len(customer_data['country'].unique()) + 1), customer_data['country'].unique(), rotation=45)
plt.show()
```
