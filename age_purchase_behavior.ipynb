```python
        #data preparation

# import package to read csv file (pandas)
import pandas as pd

# read data from csv file
customer_data = pd.read_csv("customers.csv")

```

```python
        #data exploration

age_counts = customer_data['age'].value_counts()
print(age_counts)
```

```python
        #data exploration -  grouping and aggregation
    
age_grouped = customer_data.groupby("age").agg( 
    purchase_amount_mean=("purchase_amount", "mean"),
    order_count_mean=("order_count", "mean"),
    average_order_value_mean=("average_order_value", "mean"))

age_grouped = age_grouped.round(2)

print(age_grouped)

```

```python
        #data visualization 
    
import pandas as pd
import matplotlib.pyplot as plt

# Data visualization
# Creates a new figure with a specific size of 10 inches in width and 5 inches in height
plt.figure(figsize=(10, 5))

# Bar plot for Mean Purchase Amount
plt.subplot(1, 3, 1)

plt.bar(age_grouped.index, age_grouped["purchase_amount_mean"])
plt.xlabel("Age")
plt.ylabel("Mean Purchase Amount")
plt.title("Mean Purchase Amount by Age")

# Bar plot for Mean Order Count
plt.subplot(1, 3, 2)
plt.bar(age_grouped.index, age_grouped["order_count_mean"])
plt.xlabel("Age")
plt.ylabel("Mean Order Count")
plt.title("Mean Order Count by Age")

# Bar plot for Mean Average Order Value
plt.subplot(1, 3, 3)
plt.bar(age_grouped.index, age_grouped["average_order_value_mean"])
plt.xlabel("Age")
plt.ylabel("Mean Average Order Value")
plt.title("Mean Average Order Value by Age")

plt.tight_layout()
plt.show()
```
