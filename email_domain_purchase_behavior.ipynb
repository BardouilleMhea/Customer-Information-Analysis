```python
        #data exploration 
    
import matplotlib.pyplot as plt

# Function to calculate the length of an email address
def calculate_email_length(email):
    return len(email)

# Apply the function to the 'email' column and create a new 'email_length' column
customer_data['email_length'] = customer_data['email'].apply(calculate_email_length)

# Create a histogram to visualize the distribution of email address lengths
plt.hist(customer_data['email_length'], bins=15, color='skyblue', edgecolor='black')
plt.xlabel('Email Address Length')
plt.ylabel('Frequency')
plt.title('Distribution of Email Address Lengths')
plt.show()
```
