```python
        #data visualization 
    
import pandas as pd
import matplotlib.pyplot as plt

# Group data by country and count the number of customers in each country
country_counts = customer_data['country'].value_counts()

# Create a bar plot
plt.bar(country_counts.index, country_counts.values)
plt.xlabel('Country')
plt.ylabel('Number of Customers')
plt.title('Number of Customers in Each Country')
plt.xticks(rotation=45)
plt.show()
```

```python
        #data visualization 
    
import pandas as pd
import matplotlib.pyplot as plt

# Create a pie chart
plt.pie(country_counts.values, labels=country_counts.index, autopct='%1.1f%%')
plt.title('Proportion of Customers in Each Country')
plt.axis('equal')
plt.show()
```

```python
        #data visualization 
    
import pandas as pd
import matplotlib.pyplot as plt

# Group data by country and calculate the mean purchase amount in each country
country_mean_purchase = customer_data.groupby('country')['purchase_amount'].mean()

# Create a horizontal bar plot
plt.barh(country_mean_purchase.index, country_mean_purchase.values)
plt.xlabel('Mean Purchase Amount')
plt.ylabel('Country')
plt.title('Mean Purchase Amount by Country')
plt.show()
```
