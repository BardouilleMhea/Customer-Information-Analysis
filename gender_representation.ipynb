```python
        #data visualization 
    
import pandas as pd
import matplotlib.pyplot as plt

#  Bar Plot: Gender Distribution
gender_counts = customer_data['gender'].value_counts()
plt.bar(gender_counts.index, gender_counts.values)
plt.xlabel('Gender')
plt.ylabel('Number of Customers')
plt.title('Gender Distribution Among Customers')
plt.show()
```

```python
      #data visualization 
    
import pandas as pd
import matplotlib.pyplot as plt

# Pie Chart: Gender Proportion
gender_proportions = customer_data['gender'].value_counts()
plt.pie(gender_proportions, labels=gender_proportions.index, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Gender Proportion Among Customers')
plt.show()
```

```python
      #data visualization 
    
import pandas as pd
import matplotlib.pyplot as plt

# Stacked Bar Plot: Gender Distribution by Country
gender_country_counts = customer_data.groupby(['country', 'gender']).size().unstack()
gender_country_counts.plot(kind='bar', stacked=True)
plt.xlabel('Country')
plt.ylabel('Number of Customers')
plt.title('Gender Distribution by Country')
plt.legend(title='Gender', loc='upper right')
plt.show()
```
